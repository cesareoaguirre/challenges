Consideremos  un  conjunto de caracteres compuesto por letras, un espacio (esp) y un punto (pt).  Las palabras están formadas por una o más letras,  con un máximo de veinte. El texto de entrada consiste en una o más  palabras, separadas entre sí por uno o más espacios y terminadas por cero o más espacios seguidos de un punto. Con la función de valores en caracteres LSC (Leer Siguiente Caracter), el texto de entrada debe ser leído desde e incluyendo la primera letra de la primera palabra hasta  el punto final inclusive.  Debe  producirse el texto de salida usando la primitiva ISC(x) (Imprimir Siguiente Carácter) con un parámetro valuado en caracteres. Si la función                        del programa fuera copiar el texto, el siguiente programa serviría  (suponiendo que dispusiésemos de variables valuadas en caracteres).

car  x;
repetir x:= LSC(x);  ISC hasta x = pt;

Siin embargo, en este ejemplo, el texto debe  ser sometido a la siguiente transformación:

    1) en el  texto  de salida,  las palabras sucesivas deben estar separadas por un solo espacio.
    2) en el texto de salida, la palabra final debe estar seguida por un solo punto.
    3)  cuando numeramos las palabras 0, 1, 2, 3... en el orden de izquierda  a derecha (es decir , en el que son recorridas por la evaluación repetida de LSC), las palabras con número ordinal par deben ser copiadas, mientras que las letras de las palabras con un número ordinal impar deben  ser impresas en el orden inverso.


```python
line = "Este es un programa tonto";
# input("Ingrese un planeta o escriba 'done' para terminar:")
#  Este es un programa tonto

# función recursiva  O(n)
def copyCut(phrase, index):
    print(f"phrase: {phrase} in index: {index}")
    # Se va por el último elemento del array para no tener que sobrecargar con reordenamientos
    current_word = phrase.pop(-1)
    prev_word = ""
    if len( phrase ):
        prev_word=copyCut(phrase, len( phrase )-1 )+" "

    if index>0 and index % 2 != 0:
        current_word=current_word[::-1]    
    print(f" current_word:  {current_word}")
    return f"{prev_word}{current_word}"

line_arr =  line.split(" ")
reword = copyCut(line_arr, len(line_arr)-1)


print(reword)
pass_test = "Este se un amargorp tonto"==reword
print (f"Funciona? {pass_test}")
```
